```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
x=data.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11,1),wcss)
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",max_iter=300,random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=10,color="red", label="careful")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=10,color="blue", label="standard")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=10,color="orange", label="target")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=10,color="black", label="careless")
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=10,color="magenta", label="sensible")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,color="yellow", label="centroids")
plt.xlabel("income")
plt.ylabel("spending_score")
plt.legend()
```
