
# merge join concat

```python
df1=pd.DataFrame({"name":["salman","darkk"],"no":[1,2],"val":[2,3]})
df2=pd.DataFrame({"name":["salman","darkk"],"no":[3,4],"val":[4,5]})
m=pd.merge(df1,df2,on="name")
```

```python
m
```

```python
df1=pd.DataFrame({"name":["salman","darkk"]},index=[1,2])
df2=pd.DataFrame({"just":["salman","darkk"]},index=[1,3])
df1.join(df2)      
```

```python
df1=pd.DataFrame({"name":["salman","darkk"]},index=[1,2])
df2=pd.DataFrame({"just":["salman","darkk"]},index=[1,2])
pd.concat([df1,df2],sort=True)
```
