```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
x=data.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch

dend=sch.dendrogram(sch.linkage(x,method="ward"))
plt.xlabel("Customers")
plt.ylabel("Euclidean_distance")
```

```python
from sklearn.cluster import AgglomerativeClustering 

hc=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")

y_hc=hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,color="red",label="cluster1")
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,color="blue",label="cluster2")
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,color="orange",label="cluster3")
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,color="black",label="cluster4")
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,color="magenta",label="cluster5")
plt.legend()
plt.xlabel("Customers")
plt.ylabel("score")
```
