# Importing Dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
i=pd.read_csv("iris.csv")
```

```python
i.axes
```

```python
i.plot()
```

```python
i["species"].value_counts()
```

```python
i.plot(kind="scatter",x="sepal_length",y="sepal_width")
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(i,hue="species",height=7).map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```
